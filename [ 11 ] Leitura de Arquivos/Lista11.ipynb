```python
def ler_fasta(arquivo):
    with open(arquivo, 'r') as f:
        linhas = f.readlines()
        descricao = linhas[0].strip()
        seq_DNA = ''.join([linha.strip() for linha in linhas[1:]])
    
    return {'descricao': descricao, 'seq_DNA': seq_DNA}

# Exemplo de uso:
dados = ler_fasta('Corona_genomic.fasta')
dados
```

```python
def complementar(dados):
    comp = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
    seq_complementar = ''.join([comp[nuc] for nuc in dados['seq_DNA']])
    return seq_complementar

# Exemplo de uso:
fita_complementar = complementar(dados)
print(f'Fita Inicial:      {dados["seq_DNA"]}')
print(f'Fita Complementar: {fita_complementar}')
```

```python
def transcricao(dados):
    seq_RNA = dados['seq_DNA'].replace('T', 'U')
    dados['seq_RNA'] = seq_RNA
    return dados

# Exemplo de uso:
dados_transcritos = transcricao(dados)
print(f'Fita de DNA: {dados["seq_DNA"]}')
print(f'Fita de RNA: {dados_transcritos["seq_RNA"]}')
```

```python
def freq(dados):
    seq = dados['seq_DNA']
    total = len(seq)
    frequencias = {
        'Fa': seq.count('A') / total * 100,
        'Fc': seq.count('C') / total * 100,
        'Fg': seq.count('G') / total * 100,
        'Ft': seq.count('T') / total * 100
    }
    
    dados.update(frequencias)
    
    for nuc, perc in frequencias.items():
        print(f"{nuc}: {perc:.2f}%")
    
    return dados

# Exemplo de uso:
dados_freq = freq(dados)
```

```python
# Extra: Grafico de Barra com as Frequências
import matplotlib.pyplot as plt

nucleotideos = ['A', 'C', 'G', 'T']
frequencias = [dados_freq['Fa'], dados_freq['Fc'], dados_freq['Fg'], dados_freq['Ft']]
plt.bar(nucleotideos, frequencias)
plt.ylabel('Frequência (%)')
plt.xlabel('Nucleotídeo')
plt.title('Frequência de nucleotídeos')
plt.show()
```

```python
def compare(dados1, dados2):
    seq1 = dados1['seq_DNA']
    seq2 = dados2['seq_DNA']
    
    diferencas = [(i, nuc1, nuc2) for i, (nuc1, nuc2) in enumerate(zip(seq1, seq2)) if nuc1 != nuc2]
    
    print(f"Número de nucleotídeos alterados: {len(diferencas)}")
    for pos, nuc1, nuc2 in diferencas:
        print(f"Posição {pos}: {nuc1} -> {nuc2}")

# Exemplo de uso:
dados1 = ler_fasta('seq_a.fasta')
dados2 = ler_fasta('seq_b.fasta')
compare(dados, dados2)
```

```python
import json

def salvar_json(dados, nome_arquivo):
    with open(nome_arquivo, 'w') as f:
        json.dump(dados, f, indent=4)

# Exemplo de uso:
salvar_json(dados, 'Corona_genomic.json')
```

```python
def ler_json(nome_arquivo):
    with open(nome_arquivo, 'r') as f:
        dados = json.load(f)
    return dados

# Exemplo de uso:
dados_carregados = ler_json('Corona_genomic.json')
print(dados_carregados)
```
