```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 20000
x = np.linspace(0, 1, n)
y = np.sqrt(1 - x**2)
plt.plot(x, y, color='#215a36')

# Calcular Pi com monte carlo
xx = np.random.rand(n)
yy = np.random.rand(n)
classe = (xx**2 + yy**2) < 1

plt.scatter(xx[classe], yy[classe], color='#215a36')
pi = 4 * np.sum(classe) / n
plt.title(f'Pi = {pi}')
plt.show()
```
