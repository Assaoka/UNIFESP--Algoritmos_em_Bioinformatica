# Análise de Dados com NumPy e Matplotlib

## Passo 1: Crie um array unidimensional chamado dados com 1000 elementos inteiros aleatórios no intervalo de 1 a 100.

```python
import numpy as np

dados = np.random.randint(1, 101, size=1000)
dados
```

## Passo 2: Calcule a média, mediana, desvio padrão e variância dos valores no array dados.

```python
media = np.mean(dados)
mediana = np.median(dados)
desvio_padrao = np.std(dados)
variancia = np.var(dados)

print(f"Média: {media}")
print(f"Mediana: {mediana}")
print(f"Desvio Padrão: {desvio_padrao}")
print(f"Variância: {variancia}")
```

## Passo 3: Crie um histograma com 20 bins para visualizar a distribuição dos valores no array dados.

```python
import matplotlib.pyplot as plt

plt.hist(dados, bins=20, color='blue', edgecolor='black')
plt.title('Histograma de Dados')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()
```

## Passo 4: Crie um novo array chamado dados_filtrados que contenha apenas os valores maiores que a média.


```python
dados_filtrados = dados[dados > media]
dados_filtrados
```

## Passo 5: Crie um boxplot para visualizar a distribuição dos valores no array dados_filtrados.

```python
plt.boxplot(dados_filtrados)
plt.title('Boxplot de Dados Filtrados')
plt.ylabel('Valor')
plt.show()
```

## Passo 6: Imprima as estatísticas resumidas dos valores no array dados_filtrados utilizando np.stats.describe().

```python
from scipy import stats

estatisticas = stats.describe(dados_filtrados)
print(estatisticas)
```

# Desafio 2: Processamento de Imagens Biomédicas com NumPy

## Passo 1: Importe as bibliotecas NumPy e Matplotlib.

## Passo 2: Carregue uma imagem biomédica em escala de cinza utilizando a função plt.imread().

```python
imagem = plt.imread('imagem_biomedica.png')
imagem
```

```python
imagem.shape
```

```python
# Supondo que a imagem original tenha 3 canais de cor (RGB) ou 4 (RGBA)
if len(imagem.shape) == 3:
    imagem = np.mean(imagem, axis=2)  # Converter para escala de cinza
```

```python
imagem
```

```python
imagem.shape
```

## Passo 3: Converta a imagem para um array NumPy.

```python
imagem_array = np.array(imagem)
```

## Passo 4: Normalize a intensidade dos pixels da imagem para o intervalo de 0 a 1.

```python
imagem_normalizada = imagem_array / 255.0
imagem_normalizada
```

```python
# Já estava normalizada
imagem_array.max(), imagem_array.min(), imagem_normalizada.max(), imagem_normalizada.min()
```

## Passo 5: Aplique uma transformação de negativo na imagem.

```python
imagem_negativo = 1.0 - imagem_array
```

## Passo 6: Realize uma operação de thresholding na imagem.

```python
limite = 0.5
imagem_threshold = np.where(imagem_negativo > limite, 1.0, 0.0)
```

## Passo 7: Crie um filtro de média 3x3 e aplique na imagem.

```python
filtro = np.ones((3, 3)) / 9
imagem_filtrada = np.zeros_like(imagem_threshold)

for i in range(1, imagem_threshold.shape[0] - 1):
    for j in range(1, imagem_threshold.shape[1] - 1):
        imagem_filtrada[i, j] = np.sum(imagem_threshold[i-1:i+2, j-1:j+2] * filtro)
```

## Passo 8: Imprima a imagem original, a imagem com o negativo, a imagem binarizada após o thresholding, e a imagem após o filtro de média.

```python
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(imagem_array, cmap='gray')
plt.title('Imagem Original')

plt.subplot(2, 2, 2)
plt.imshow(imagem_negativo, cmap='gray')
plt.title('Imagem Negativa')

plt.subplot(2, 2, 3)
plt.imshow(imagem_threshold, cmap='gray')
plt.title('Imagem Binarizada')

plt.subplot(2, 2, 4)
plt.imshow(imagem_filtrada, cmap='gray')
plt.title('Imagem Filtrada')

plt.tight_layout()
plt.show()
```
